```python
dfr = pd.DataFrame()
for col in df.columns:    
    if is_numeric_dtype(df[col]):
        plot = df[col].hist()
        file = os.path.join(path,col.replace(" ","")+'.png')
        plt.savefig(file)
        data = {"Column":col, "Type":df.dtypes[col], "Min":df[col].min(),"Average":df[col].mean(), "Max":df[col].max(), "Median":df[col].median(), "Standard Deviation":df[col].std(), "Histogram":file}
        dfr = dfr.append(data, ignore_index=True)
    elif is_string_dtype(df[col]):
        data = {"Column":col, "Type":df.dtypes[col], "Unique":len(df[col].unique()), "NaNs":df[col].isna().sum()}
        dfr = dfr.append(data, ignore_index=True)
        
    elif is_datetime(df[col]):
        data = {"Column":col, "Type":df.dtypes[col], "First Date":df[col].min().date().strftime("%d %b %Y"), "Last Date":df[col].max().date().strftime("%d %b %Y")}
        dfr = dfr.append(data, ignore_index=True)        
        
dfr = dfr[['Column','Type', 'Unique', 'NaNs','Min','Average','Max','Median','Standard Deviation', "Histogram"]]
```
